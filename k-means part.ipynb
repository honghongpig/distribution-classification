```python
import pandas as pd

# 读取 Excel 文件（如果是 .xls 格式也适用）
file_path = "F:/distrabution clasification/customers_with_address_complete_deduplicated_29868_osmid.csv"  # 替换成你的 Excel 路径
data = pd.read_csv(file_path)

# 显示前 5 行
data.head()
```

```python
# 筛选出 station_name == "太原直送" 的行，命名为 data1
data1 = data[data["station_name"] == "太原直送"]

# 查看筛选结果（显示前5行）
print("筛选后的数据量：", len(data1))  # 查看有多少行符合条件
data1.head()
```

```python
# 1. 查看数据形状（行数、列数）
print("数据形状（行, 列）：", data1.shape)  # 例如 (29868, 10) 表示29868行10列

# 2. 查看列名和数据类型，以及缺失值数量
print("\n数据基本信息：")
data1.info()
```

```python
#检查经纬度缺失值
data1 = data1.dropna(subset=["longitude", "latitude"])
```

```python
data1 = data1.drop(columns=["street"])
```

```python
#检查经纬度范围
print("\n经纬度范围检查：")
print("纬度最小值：", data["latitude"].min())
print("纬度最大值：", data["latitude"].max())
print("经度最小值：", data["longitude"].min())
print("经度最大值：", data["longitude"].max())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans  # 导入K-means

# 设置中文显示
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]


X = data1[["longitude", "latitude"]].values  # 经纬度数据

# 2. K-means 分块（指定簇数，如10块，可调整）
n_clusters = 2  # 分块数量（建议5-20，根据密集程度调整）
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
data1["kmeans_cluster"] = kmeans.labels_  # 为所有点添加簇标签

print(f"保留所有数据点，共 {len(data1)} 行，分为 {n_clusters} 块")

# 3. 可视化所有点的分块结果
plt.figure(figsize=(12, 10))

# 绘制每个簇的所有点（用不同颜色区分块）
for cluster_id in range(n_clusters):
    # 提取该簇的所有点（不采样，全部保留）
    cluster_all = data1[data1["kmeans_cluster"] == cluster_id]
    plt.scatter(
        cluster_all["longitude"],
        cluster_all["latitude"],
        s=5,  # 点大小设小，避免密集区域重叠
        alpha=0.6,  # 适当透明，密集区域也能看清层次
        label=f"块 {cluster_id}",
        edgecolors="black",
        linewidths=0.2  # 边缘线细一点，减少视觉干扰
    )

# 叠加K-means簇中心（红色五角星，代表每个块的核心）
cluster_centers = kmeans.cluster_centers_
plt.scatter(
    cluster_centers[:, 0],  # 中心经度
    cluster_centers[:, 1],  # 中心纬度
    s=200,
    marker="*",
    c="red",
    label="块中心",
    edgecolors="black",
    linewidths=1
)

# 聚焦核心区域
plt.xlim(111.5, 113.0)
plt.ylim(37.5, 38.5)
plt.xlabel("经度", fontsize=12)
plt.ylabel("纬度", fontsize=12)
plt.title(f"K-means分块结果（保留所有点，共{n_clusters}块）", fontsize=15)
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")  # 图例放右侧，避免遮挡点
plt.grid(linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import os  # 用于处理文件路径

# 按簇筛选并保存为CSV到指定路径
save_path = "F:/distrabution clasification"  # 目标保存路径

# 检查路径是否存在，不存在则创建
if not os.path.exists(save_path):
    os.makedirs(save_path)  # 自动创建文件夹（包括多级目录）

for cluster_id in range(n_clusters):
    # 筛选当前簇的所有数据
    cluster_data = data1[data1["kmeans_cluster"] == cluster_id]
    # 拼接完整保存路径（路径+文件名）
    file_name = f"cluster_{cluster_id}.csv"
    full_path = os.path.join(save_path, file_name)  # 组合路径和文件名
    
    # 保存为CSV
    cluster_data.to_csv(full_path, index=False, encoding="utf-8-sig")
    print(f"簇 {cluster_id} 已保存到：{full_path}，共 {len(cluster_data)} 行")
```

对簇0进行DBSCAN

```python
from sklearn.cluster import DBSCAN


# 1. 读取数据
file_path = "F:/distrabution clasification/cluster_0.csv"
cluster0_data = pd.read_csv(file_path)
X_cluster0 = cluster0_data[["longitude", "latitude"]].values

# 2. DBSCAN聚类（保持原参数）
dbscan = DBSCAN(eps=0.01, min_samples=30).fit(X_cluster0)
cluster0_data["dbscan_label"] = dbscan.labels_
labels = dbscan.labels_

# 3. 手动精准锁定核心密集区（基于放大后的视觉范围）
lon_lower, lon_upper = 112.0, 113.0
lat_lower, lat_upper = 37.5, 38.2

# 4. 可视化（仅显示核心区，点自然均匀分布）
plt.figure(figsize=(10, 8))
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for label, color in zip(unique_labels, colors):
    if label == -1:
        color = [0.5, 0.5, 0.5, 0.5]
    plt.scatter(
        X_cluster0[labels == label, 0],
        X_cluster0[labels == label, 1],
        s=8,
        alpha=0.7,
        color=color,
        edgecolors="black",
        linewidths=0.3,
        label=f"子簇 {label}" if label != -1 else "噪声点"
    )

# 应用精准范围，彻底裁掉空白
plt.xlim(lon_lower, lon_upper)
plt.ylim(lat_lower, lat_upper)

plt.xlabel("经度", fontsize=12)
plt.ylabel("纬度", fontsize=12)
plt.title("cluster_0 的 DBSCAN 分布", fontsize=15)
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#检查经纬度范围
print("\n经纬度范围检查：")
print("纬度最小值：", cluster0_data["latitude"].min())
print("纬度最大值：", cluster0_data["latitude"].max())
print("经度最小值：", cluster0_data["longitude"].min())
print("经度最大值：", cluster0_data["longitude"].max())
```

```python
# 3. 统计每一簇的点数
# 使用value_counts()统计每个标签的出现次数
cluster_counts = pd.Series(labels).value_counts().sort_index()

# 打印结果（区分正常簇和噪声点）
print("各簇点数统计：")
for label, count in cluster_counts.items():
    if label == -1:
        print(f"噪声点：{count} 个")
    else:
        print(f"子簇 {label}：{count} 个")
```

```python
# 4. 筛选点数小于100的簇并输出为CSV，文件名包含点数
output_dir = "F:/distrabution clasification/"  # 输出文件夹路径（可根据需要修改）
small_clusters = []

for label, count in cluster_counts.items():
    if label != -1 and count < 100:  # 排除噪声点，只处理点数<100的正常簇
        # 提取当前簇的数据
        cluster_data = cluster0_data[cluster0_data["dbscan_label"] == label]
        # 定义输出文件名（格式：簇标签_点数.csv）
        output_filename = f"cluster0_subcluster_{label}_points_{count}.csv"
        output_path = output_dir + output_filename
        # 保存为CSV
        cluster_data.to_csv(output_path, index=False)
        small_clusters.append((label, count))
        print(f"已保存点数小于100的簇：{output_filename}")

if not small_clusters:
    print("没有点数小于100的簇")
```

```python
# 提取子簇4的数据（基于之前DBSCAN的标签）
subcluster4_data = cluster0_data[cluster0_data["dbscan_label"] == 4].copy()
X_sub4 = subcluster4_data[["longitude", "latitude"]].values

# 检查子簇4是否存在
if len(subcluster4_data) == 0:
    print("子簇4不存在，请确认簇标签是否正确")
else:
    print(f"子簇4包含 {len(subcluster4_data)} 个数据点，开始K-means聚类（分为2类）...")
 # 2. 对子簇4执行K-means聚类（固定分为2类）
    kmeans = KMeans(n_clusters=2, random_state=42)  # random_state确保结果可复现
    subcluster4_data["kmeans_label"] = kmeans.fit_predict(X_sub4)
    kmeans_labels = subcluster4_data["kmeans_label"].values

    # 3. 统计2个子类的点数
    kmeans_counts = pd.Series(kmeans_labels).value_counts().sort_index()
    print("\nK-means聚类（2类）结果统计：")
    for label in [0, 1]:
        print(f"子类 {label}：{kmeans_counts.get(label, 0)} 个数据点")

    # 4. 可视化聚类结果
    plt.figure(figsize=(10, 8))
    # 定义两类的颜色（可自定义）
    colors = {0: 'skyblue', 1: 'orange', -1: 'gray'}
    
    for label in [0, 1]:
        # 筛选当前类别的点
        mask = kmeans_labels == label
        plt.scatter(
            X_sub4[mask, 0],  # 经度
            X_sub4[mask, 1],  # 纬度
            s=15,
            alpha=0.8,
            color=colors[label],
            edgecolors="black",
            linewidths=0.3,
            label=f"子类 {label}（{kmeans_counts[label]}点）"
        )

    # 标记聚类中心
    centers = kmeans.cluster_centers_
    plt.scatter(
        centers[:, 0],  # 中心经度
        centers[:, 1],  # 中心纬度
        s=200,
        marker='X',
        c='red',
        label='聚类中心'
    )

    # 自动调整坐标范围，聚焦子簇4区域
    plt.xlim(X_sub4[:, 0].min() - 0.01, X_sub4[:, 0].max() + 0.01)
    plt.ylim(X_sub4[:, 1].min() - 0.01, X_sub4[:, 1].max() + 0.01)

    plt.xlabel("经度", fontsize=12)
    plt.ylabel("纬度", fontsize=12)
    plt.title("子簇4的K-means聚类结果（分为2类）", fontsize=15)
    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.grid(linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import silhouette_score
# 提取子簇14的数据
subcluster14_data = cluster0_data[cluster0_data["dbscan_label"] == 14].copy()
X_sub14 = subcluster14_data[["longitude", "latitude"]].values
total_points = len(subcluster14_data)

# 检查子簇14的合法性
if total_points == 0:
    print("子簇14不存在，请确认簇标签是否正确")
elif total_points < 2:
    print(f"子簇14仅包含{total_points}个点，无法进行聚类")
else:
    print(f"子簇14包含{total_points}个点，开始自动确定聚类数量...")

    # 2. 计算最大可能的聚类数量（确保每类<100）
    max_clusters = min(total_points // 1, total_points // 1)  # 至少1点/类
    max_clusters = min(max_clusters, total_points // 1)  # 实际最大聚类数（每类至少1点）
    max_clusters = min(max_clusters, 10)  # 限制最大聚类数（避免过多类）
    if max_clusters < 2:
        max_clusters = 2  # 至少分2类

    # 确保每类点数<100
    max_possible_clusters = (total_points + 99) // 100  # 向上取整（如150点→2类，250点→3类）
    max_clusters = max(max_clusters, max_possible_clusters)  # 保证每类<100
    max_clusters = min(max_clusters, total_points)  # 防止超过总点数

    print(f"根据总点数，可能的聚类数量范围：2~{max_clusters}（确保每类<100）")

    # 3. 计算不同聚类数的轮廓系数，选择最优k
    best_k = 2
    best_silhouette = -1
    silhouette_scores = []
    possible_ks = range(2, max_clusters + 1)

    for k in possible_ks:
        kmeans = KMeans(
            n_clusters=k,
            init='k-means++',
            n_init=30,
            random_state=42
        )
        labels = kmeans.fit_predict(X_sub14)
        # 计算轮廓系数（值越接近1越好）
        sil_score = silhouette_score(X_sub14, labels)
        silhouette_scores.append(sil_score)
        # 检查是否为最优k
        if sil_score > best_silhouette:
            best_silhouette = sil_score
            best_k = k

    print(f"最优聚类数量：{best_k}（轮廓系数：{best_silhouette:.3f}）")

    # 4. 使用最优k进行最终聚类
    kmeans = KMeans(
        n_clusters=best_k,
        init='k-means++',
        n_init=50,
        max_iter=1000,
        random_state=42
    )
    subcluster14_data["kmeans_label"] = kmeans.fit_predict(X_sub14)
    kmeans_labels = subcluster14_data["kmeans_label"].values

    # 5. 验证聚类结果（每类<100且均衡）
    kmeans_counts = pd.Series(kmeans_labels).value_counts().sort_index()
    max_count = kmeans_counts.max()
    min_count = kmeans_counts.min()
    balance_ratio = min_count / max_count if max_count != 0 else 0

    # 检查每类是否<100
    if max_count >= 100:
        print(f"警告：聚类后最大类包含{max_count}点，超过100点，请增大聚类数量")
    else:
        print(f"聚类结果满足每类点数<100的条件（最大类：{max_count}点）")

    # 打印统计信息
    print("\nK-means聚类结果统计：")
    for label, count in kmeans_counts.items():
        print(f"子类 {label}：{count} 个数据点")
    print(f"均衡比例：{balance_ratio:.2f}（接近1表示更均衡）")

    # 6. 输出各子类数据（文件名包含点数）
    output_dir = "F:/distrabution clasification/"
    for label in kmeans_counts.index:
        subclass_data = subcluster14_data[subcluster14_data["kmeans_label"] == label]
        point_count = len(subclass_data)
        output_filename = f"subcluster14_subclass_{label}_points_{point_count}.csv"
        output_path = output_dir + output_filename
        subclass_data.to_csv(output_path, index=False)
        print(f"已保存：{output_filename}")

    # 7. 可视化聚类结果
    plt.figure(figsize=(10, 8))
    # 生成足够的颜色
    colors = plt.cm.Set3(np.linspace(0, 1, best_k))

    for label in range(best_k):
        mask = kmeans_labels == label
        plt.scatter(
            X_sub14[mask, 0],
            X_sub14[mask, 1],
            s=15,
            alpha=0.8,
            color=colors[label],
            edgecolors="black",
            linewidths=0.3,
            label=f"子类 {label}（{kmeans_counts[label]}点）"
        )

    # 标记聚类中心
    centers = kmeans.cluster_centers_
    plt.scatter(
        centers[:, 0],
        centers[:, 1],
        s=200,
        marker='X',
        c='black',
        label='聚类中心'
    )

    # 聚焦子簇14的坐标范围
    plt.xlim(X_sub14[:, 0].min() - 0.01, X_sub14[:, 0].max() + 0.01)
    plt.ylim(X_sub14[:, 1].min() - 0.01, X_sub14[:, 1].max() + 0.01)

    plt.xlabel("经度", fontsize=12)
    plt.ylabel("纬度", fontsize=12)
    plt.title(f"子簇14的K-means聚类结果（{best_k}类，均衡比例：{balance_ratio:.2f}）", fontsize=15)
    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.grid(linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.show()

    # 可选：可视化轮廓系数选择过程
    plt.figure(figsize=(8, 5))
    plt.plot(possible_ks, silhouette_scores, 'bo-')
    plt.xlabel("聚类数量k")
    plt.ylabel("轮廓系数")
    plt.title("不同k值的轮廓系数（越高越好）")
    plt.grid(linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# 5. 提取簇6数据并进行K-means细分
cluster6_data = cluster1_data[cluster1_data["dbscan_label"] == 6].copy()
total_points_cluster6 = len(cluster6_data)

# 检查簇6是否存在及规模
if total_points_cluster6 == 0:
    print("\n簇6不存在，请确认DBSCAN聚类结果")
elif total_points_cluster6 < 2:
    print(f"\n簇6仅包含{total_points_cluster6}个点，无法进行K-means聚类")
else:
    print(f"\n提取到簇6，包含 {total_points_cluster6} 个点，开始K-means细分...")
    X_cluster6 = cluster6_data[["longitude", "latitude"]].values

    # 计算合理的聚类数量（确保每类<100且均衡）
    max_per_cluster = 100  # 每类最大点数限制
    min_clusters = max(2, (total_points_cluster6 + max_per_cluster - 1) // max_per_cluster)  # 最小聚类数
    max_possible_clusters = total_points_cluster6 // 20  # 最大聚类数（避免类过小）
    n_clusters = min(min_clusters, max_possible_clusters)
    n_clusters = max(n_clusters, 2)  # 至少分2类
    print(f"根据簇6规模，K-means聚类数量设为：{n_clusters}（确保每类<100点）")

    # 执行K-means聚类（优化均衡性）
    kmeans_cluster6 = KMeans(
        n_clusters=n_clusters,
        init='k-means++',
        n_init=50,
        random_state=42
    )
    cluster6_data["kmeans_label"] = kmeans_cluster6.fit_predict(X_cluster6)
    kmeans_labels_cluster6 = cluster6_data["kmeans_label"].values

    # 统计并验证结果
    kmeans_counts_cluster6 = pd.Series(kmeans_labels_cluster6).value_counts().sort_index()
    max_count = kmeans_counts_cluster6.max()
    min_count = kmeans_counts_cluster6.min()
    balance_ratio = min_count / max_count if max_count != 0 else 0

    # 检查每类是否<100
    if max_count >= 100:
        print(f"警告：最大子类包含{max_count}点，超过100点，建议将聚类数量增至{n_clusters + 1}")
    else:
        print(f"聚类结果满足每类点数<100的条件（最大类：{max_count}点）")

    # 打印统计信息
    print("\nK-means细分结果统计：")
    for label, count in kmeans_counts_cluster6.items():
        print(f"子类 {label}：{count} 个点")
    print(f"均衡比例：{balance_ratio:.2f}（接近1表示更均衡）")
      # 可视化聚类结果
    plt.figure(figsize=(10, 8))
    colors = plt.cm.Set1(np.linspace(0, 1, n_clusters))  # 用Set1色系与其他簇区分

    for label in range(n_clusters):
        mask = kmeans_labels_cluster6 == label
        plt.scatter(
            X_cluster6[mask, 0],  # 经度
            X_cluster6[mask, 1],  # 纬度
            s=15,
            alpha=0.8,
            color=colors[label],
            edgecolors="black",
            linewidths=0.3,
            label=f"子类 {label}（{kmeans_counts_cluster6[label]}点）"
        )

    # 标记聚类中心
    centers = kmeans_cluster6.cluster_centers_
    plt.scatter(
        centers[:, 0],
        centers[:, 1],
        s=200,
        marker='X',
        c='green',
        label='聚类中心'
    )

    # 聚焦簇6的坐标范围（结合之前的0.02度≈2.23公里，适当扩展范围）
    plt.xlim(X_cluster6[:, 0].min() - 0.01, X_cluster6[:, 0].max() + 0.01)  # 约±1.1公里
    plt.ylim(X_cluster6[:, 1].min() - 0.01, X_cluster6[:, 1].max() + 0.01)

    plt.xlabel("经度", fontsize=12)
    plt.ylabel("纬度", fontsize=12)
    plt.title(f"簇6的K-means细分结果（{n_clusters}类，均衡比例：{balance_ratio:.2f}）", fontsize=15)
    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.grid(linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.show()
```
